```python
import pandas as pd

# Load the CSV file to examine the data
file_path = "req_cert.csv"
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
df.head()
```

```python
from datetime import datetime

# Set current date for comparison
current_date = datetime.now()

# Create a dictionary to store results
cert_status = {"Upcoming": [], "Overdue": []}

# Loop through each row and check validity of each certificate
for index, row in df.iterrows():
    for col in df.columns:
        if 'Validity' in col and isinstance(row[col], str) and row[col].strip():
            # Convert the date string to a datetime object
            cert_date = datetime.strptime(row[col], '%Y-%m-%d %H:%M:%S')
            # Check if the certification date is upcoming or overdue
            if cert_date > current_date:
                cert_status["Upcoming"].append({
                    "Employee": row['Full Name'],
                    "Certification": col,
                    "Date": cert_date,
                    "days_remaining":(cert_date-current_date).days
                })
            else:
                cert_status["Overdue"].append({
                    "Employee": row['Full Name'],
                    "Certification": col,
                    "Date": cert_date,
                    "over_due_days":(current_date-cert_date).days
                })

# Convert results to DataFrame for better readability
upcoming_df = pd.DataFrame(cert_status["Upcoming"])
overdue_df = pd.DataFrame(cert_status["Overdue"])

upcoming_df, overdue_df
```

```python
overdue_df.to_csv("overdue_df.csv")
```

```python
upcoming_df.to_csv("upcoming_df.csv")
```
